# Definición del problema

# Precio del oro 

El precio del oro ha sido un reflejo crucial de las dinámicas económicas y políticas globales a lo largo de la historia. Desde su papel central en el sistema monetario bajo el patrón oro hasta su fluctuación libre en los mercados actuales, el oro ha servido como un indicador de estabilidad y un refugio seguro en tiempos de crisis. Eventos históricos como la Gran Depresión, el colapso del sistema de Bretton Woods en 1971, la crisis del petróleo de los años 70, la crisis financiera de 2008 y la pandemia de COVID-19 han sido periodos atípicos que marcaron picos en su valor. Estos aumentos reflejan una tendencia donde, en tiempos de incertidumbre económica y geopolítica, la demanda de oro como activo seguro se dispara. El precio del oro está influido por variables exógenas como la inflación, las tasas de interés, el dólar estadounidense y la estabilidad política. Estos factores externos, junto con las percepciones de riesgo, determinan el comportamiento del mercado del oro. Al mismo tiempo, el oro también impacta otras áreas de la economía, actuando como una variable endógena que influye en las políticas monetarias y en las decisiones de inversión. Pronosticar el precio del oro es de gran importancia para inversores, bancos centrales, gobiernos y empresas que dependen de este recurso. Un pronóstico preciso permite tomar decisiones informadas sobre la gestión del riesgo, la política económica y la asignación de recursos. Dado su papel como un barómetro de la confianza del mercado y las expectativas inflacionarias, el análisis del comportamiento del oro no solo ayuda a proteger el capital, sino que también ofrece una comprensión más profunda de las fuerzas económicas globales que lo impulsan.

# Inflación

La inflación, como indicador del aumento generalizado de precios en una economía, ha jugado un papel crucial en la historia económica mundial, reflejando la interacción entre políticas monetarias, choques económicos y dinámicas sociales. A lo largo de las décadas, la inflación ha experimentado fluctuaciones notables, desde la alta inflación de los años 70 y 80, impulsada por los shocks petroleros, hasta los recientes repuntes post-crisis financiera de 2008 y la pandemia de COVID-19. Estos periodos atípicos han llevado a cambios significativos en la manera en que los bancos centrales y los gobiernos manejan las políticas monetarias y fiscales, buscando siempre controlar y prever las tendencias inflacionarias.

La inflación es afectada por variables exógenas como los precios de las materias primas, las políticas monetarias, el tipo de cambio, y las condiciones del mercado laboral. Estas variables externas pueden desencadenar presiones inflacionarias, que a su vez influyen en las decisiones de política económica, afectando el crecimiento económico y la estabilidad social. Al mismo tiempo, la inflación actúa como una variable endógena que condiciona las políticas públicas, la inversión y la confianza del consumidor.

Pronosticar la inflación es de vital importancia para actores económicos clave, incluidos los bancos centrales, los gobiernos, los inversores y las empresas. Para los bancos centrales, prever la inflación es fundamental para diseñar políticas que mantengan la estabilidad de precios y promuevan el crecimiento económico. Los gobiernos utilizan estas previsiones para ajustar políticas fiscales y garantizar el bienestar económico. Los inversores y las empresas, por su parte, dependen de estas proyecciones para tomar decisiones estratégicas en cuanto a precios, inversiones y planificación financiera. En definitiva, el análisis y pronóstico de la inflación no solo son esenciales para la estabilidad económica, sino también para el bienestar general a largo plazo.

# Transformaciones y Ajustes 1

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Descargar los datos
gold_symbol = 'GC=F'  # Contrato de futuros del oro
start_date = '2020-01-01'
end_date = '2024-08-23'
gold_data = yf.download(gold_symbol, start=start_date, end=end_date)

# Primeros 5 registros
print(gold_data.head())

# Grafico inicial
plt.figure(figsize=(12,6))
plt.plot(gold_data['Close'], label='Precio de cierre')
plt.title('Precio del Oro (2020 - 2024)')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()

# Paso 1: Interpolación de valores faltantes
gold_data['Close'] = gold_data['Close'].interpolate()

# Paso 2: Transformación logarítmica
gold_data['Log_Close'] = np.log(gold_data['Close'])

# Gráfico de la serie transformada (logarítmica)
plt.figure(figsize=(12,6))
plt.plot(gold_data['Log_Close'], label='Precio de cierre (Log)')
plt.title('Precio del Oro Transformado (Log)')
plt.xlabel('Fecha')
plt.ylabel('Log(Precio en USD)')
plt.legend()
plt.show()

# Paso 3: Descomposición de la serie temporal (tendencia, estacionalidad, residuales)
result = seasonal_decompose(gold_data['Log_Close'], model='additive', period=365)

# Graficar la descomposición
result.plot()
plt.show()

# Paso 4: Ajuste de un modelo simple de regresión lineal para predecir tendencias
# Crear variable de tiempo como índice para entrenar el modelo
gold_data['Time'] = np.arange(len(gold_data.index))

# Dividir datos en entrenamiento y prueba
train_data = gold_data.iloc[:-100]  # Entrenamos con todos menos los últimos 100 días
test_data = gold_data.iloc[-100:]  # Prueba con los últimos 100 días

X_train = train_data[['Time']]
y_train = train_data['Log_Close']
X_test = test_data[['Time']]
y_test = test_data['Log_Close']

# Entrenar el modelo
model = LinearRegression()
model.fit(X_train, y_train)

# Predicciones
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Evaluación del modelo
train_mse = mean_squared_error(y_train, train_predictions)
test_mse = mean_squared_error(y_test, test_predictions)

print(f'Error cuadrático medio (MSE) en entrenamiento: {train_mse}')
print(f'Error cuadrático medio (MSE) en prueba: {test_mse}')

# Graficar predicciones vs datos reales en el conjunto de prueba
plt.figure(figsize=(12,6))
plt.plot(test_data.index, y_test, label='Datos Reales (Log)')
plt.plot(test_data.index, test_predictions, label='Predicción (Regresión Lineal)', linestyle='--')
plt.title('Predicciones de precios del oro (logarítmico)')
plt.xlabel('Fecha')
plt.ylabel('Log(Precio en USD)')
plt.legend()
plt.show()

# Paso 5: Revertir la transformación logarítmica para obtener precios originales
test_predictions_original = np.exp(test_predictions)

plt.figure(figsize=(12,6))
plt.plot(test_data.index, np.exp(y_test), label='Datos Reales')
plt.plot(test_data.index, test_predictions_original, label='Predicción (Original)', linestyle='--')
plt.title('Predicción del precio del oro (Original)')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()
```

# Especificación del modelo

Observando las graficas anteriores podemos analizar como el precio del oro puede ser obeservado como de manera volatil pero en si va en una tendencia de incremento. Cuando esta baja de precio se puede calcular como un fallo o error, pero este igual brnda función para entender hacia donde va esta. 

# Estimación y evaluación de residuales y error

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error

# Predicciones sobre el conjunto de prueba (usando el modelo de regresión lineal ya entrenado)
y_test = test_data['Log_Close']
test_predictions = model.predict(X_test)

# Paso 1: Cálculo de los residuales
residuals = y_test - test_predictions

# Paso 2: Evaluación de los residuales
plt.figure(figsize=(12,6))
plt.plot(test_data.index, residuals, label='Residuales')
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuales de las predicciones')
plt.xlabel('Fecha')
plt.ylabel('Residuales (Log(Precio))')
plt.legend()
plt.show()

# Histograma de los residuales para verificar la distribución
plt.figure(figsize=(12,6))
plt.hist(residuals, bins=20, edgecolor='black')
plt.title('Distribución de los residuales')
plt.xlabel('Residuales')
plt.ylabel('Frecuencia')
plt.show()

# Paso 3: Cálculo del error en el conjunto de prueba
mse = mean_squared_error(y_test, test_predictions)
rmse = np.sqrt(mse)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Si quieres visualizar los residuales en términos del precio original:
# Revertir la transformación logarítmica
residuals_original = np.exp(y_test) - np.exp(test_predictions)

plt.figure(figsize=(12,6))
plt.plot(test_data.index, residuals_original, label='Residuales (Precio original)')
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuales de las predicciones (Precio original)')
plt.xlabel('Fecha')
plt.ylabel('Residuales (USD)')
plt.legend()
plt.show()
```

# Comparar modelos

Los residuales son la diferencia entre los valores reales y las predicciones del modelo. Se evalúan para verificar la aleatoriedad y asegurarse de que no haya patrones detectables, lo que indicaría que el modelo ha capturado correctamente la tendencia.

De la descomposición: Se lleva a cabo por medio de la interpolación, la transformación logaritmica, la regresión lineal y la reversión de la transformación logaritmica. 

# Pronostico

```python
# Paso 1: Extender el rango temporal para pronósticos futuros
future_days = 30  # Pronosticar los próximos 30 días
last_time_index = gold_data['Time'].iloc[-1]
future_time = np.arange(last_time_index + 1, last_time_index + future_days + 1).reshape(-1, 1)

# Paso 2: Realizar las predicciones futuras utilizando el modelo entrenado
future_predictions_log = model.predict(future_time)

# Convertir las predicciones logarítmicas a la escala original
future_predictions_original = np.exp(future_predictions_log)

# Paso 3: Crear un rango de fechas para los pronósticos
future_dates = pd.date_range(start=gold_data.index[-1] + pd.Timedelta(days=1), periods=future_days)

# Paso 4: Visualizar el pronóstico junto con los datos reales
plt.figure(figsize=(12,6))
plt.plot(gold_data.index, gold_data['Close'], label='Datos Reales (Histórico)')
plt.plot(future_dates, future_predictions_original, label='Pronóstico', linestyle='--')
plt.title('Pronóstico del Precio del Oro para los Próximos 30 Días')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.legend()
plt.show()

# Mostrar las predicciones futuras
future_forecast = pd.DataFrame({'Fecha': future_dates, 'Pronóstico': future_predictions_original})
print(future_forecast)
```
